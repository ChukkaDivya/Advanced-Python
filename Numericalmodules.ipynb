```python
from random import randint
```

```python
marks=[randint(50,100) for _ in range(10)]
```

```python
marks
```

```python
import statistics
```

```python
avg_marks=statistics.mean(marks)
```

```python
avg_marks
```

```python
highest_marks=max(marks)
```

```python
highest_marks
```

```python
import math
```

```python
performance=math.sqrt((avg_marks**2+highest_marks**2)/2)
```

```python
performance
```
