```python
class Calculator:
    def __init__(self,num1,num2):
        self.num1=num1
        self.num2=num2
    def __add__(self,other):
        return Calculator(self.num1+other.num1,self.num2+other.num2)
    def __mul__(self,other):
        return Calculator(self.num1*other.num1,self.num2*other.num2)
    def __sub__(self,other):
        return Calculator(self.num1-other.num1,self.num2-other.num2)
    def __truediv__(self,other):
        print("Division")
        return Calculator(self.num1/other.num1,self.num2/other.num2)
    def __floordiv__(self,other):
        print("Floor Division")
        return Calculator(self.num1//other.num1,self.num2//other.num2)
    def __str__(self):
        return f"Calculates values:{self.num1} and {self.num2}"
```

```python
op_1=Calculator(10,20)
```

```python
op_2=Calculator(5,10)
```

```python
op_3=Calculator(2,5)
```

```python
print(op_1+op_2+op_3)
```

```python
print(op_1*op_2*op_3)
```

```python
print(op_1-op_2-op_3)
```

```python
print(op_1/op_2)
```

```python
print(op_2//op_3)
```

```python
class Comparision:
    def __init__(self,num1,num2):
        self.num1=num1
        self.num2=num2
    def __gt__(self,other):
        return self.num1>other.num1,self.num2>other.num2
    def __lt__(self,other):
        return self.num1<other.num1,self.num2<other.num2
    def __ge__(self,other):
        return self.num1>=other.num1,self.num2>=other.num2   
    def __le__(self,other):
        return self.num1<=other.num1,self.num2<=other.num2
    def __eq__(self,other):
        return self.num1==other.num1,self.num2==other.num2
    def __ne__(self,other):
        return self.num1!=other.num1,self.num2!=other.num2
```

```python
c1=Comparision(100,50)
```

```python
c2=Comparision(20,20)
```

```python
c3=Comparision(5,20)
```

```python
print(c1>c2>c3)
```

```python
print(c1<c2<c3)
```

```python
print(c1==c2==c3)
```

```python
print(c1>=c2>=c3)
```

```python
print(c1<=c2<=c3)
```

```python
print(c1>c2!=c3)
```

```python
class Container:
    def __init__(self,var):
        self.var=var
    def __len__(self):
        return len(self.var)
    def __getitem__(self,index):
        return self.var[index]
    def __setitem__(self,index,value):
        self.var[index]=value
    def __delitem__(self,index):
        del self.var[index]
    def __contains__(self,value):
        return value in self.var
    def __iter__(self):
        return iter(self.var)
```

```python
cd=Container([1,2,3,4,5])
```

```python
print("length of the list is",len(cd))
```

```python
print("Item at index 4 is",cd[4])
```

```python
cd[4]=6
```

```python
print('updating item at index 4',cd[4])
```

```python
del(cd[4])
```

```python
len(cd)
```

```python
print('Checking list item',6 not in cd)
```

```python
for i in cd:
    print(i,end=" ")
```
