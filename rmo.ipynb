```python
class Base:
    def show(self):
        print("base class")
class A:
    def show(self):
        print("class A")
        super().show()
class B:
    def show(self):
        print("class B")
        super().show()
class C(A,B,Base):
    def show(self):
        super().show()
        print("class C")
```

```python
class Base:
    def __init__(self,*args):
        print(f"base")
    def show(self):
        print("base class")
class A:
    def __init__(self,a,*args):
        self.a=a
        print(f"Class A {self.a}")
        super().__init__(*args)
    def show(self):
        print("class A")
        super().show()
class B:
    def __init__(self,b,*args):
        self.b=b
        print(f"Class B {self.b}")
        super().__init__(*args)
    def show(self):
        print("class B")
        super().show()
class C(A,B,Base):
    def __init__(self,a,b):
        super().__init__(a,b)
        print(f"C class {self.a} and {self.b}")
    def show(self):
        super().show()
        print("class C")
```

```python
obj_4=C(4,5)
```

```python
class Base:
    def __init__(self):
        print(f"base")
    def show(self):
        print("base class")
class A:
    def __init__(self,a):
        self.a=a
        print(f"Class A {self.a}")
    def show(self):
        print("class A")
        super().show()
class B:
    def __init__(self,b):
        self.b=b
        print(f"Class B {self.b}")
    def show(self):
        print("class B")
        super().show()
class C(A,B,Base):
    def __init__(self,a,b):
        A.__init__(self,a)
        B.__init__(self,b)
        print(f"C class {self.a} and {self.b}")
    def show(self):
        super().show()
        print("class C")
```

```python
obj_5=C(4,5)
```
