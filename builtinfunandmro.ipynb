```python
class Animal:
    def __init__(self,name):
        self.name=name
    def eat(self):
        print('eating')
    def sound(self):
        pass
class Dog(Animal):
    def __init__(self,name,breed):
        super().__init__(name)
        self.breed=breed
    def sound(self):
        print(f'{self.name} says bow bow')
```

```python
dog1=Dog('Bruno','husky')
dog1.sound()
```

```python
#built in functions
print(type(dog1))
```

```python
print(issubclass(Dog,Animal))
```

```python
class A:
    def show(self):
        super().show()
        print("class A")
class B:
    def show(self):
        print("class B")
class C(A,B):
    def show(self):
        super().show()
        print("class C")
        A.show(self)
```

```python
obj=C()
```

```python
obj.show()
```
