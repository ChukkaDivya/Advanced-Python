### Single inheritance

```python
def eat(self):
    print("eating")
def sound(self):
    pass
```

```python
Animal=type('Animal',(),{'eat':eat,'sound':sound})
```

```python
def __dinit__(self,name):
    self.name=name
def speak(self):
    print(f"{self.name} says bow bow !")
```

```python
Dog=type('Dog',(Animal,),{'__init__':__dinit__,'sound':speak})
```

```python
single=Dog('Buddy')
```

```python
single.sound()
single.eat()
```

### Multiple inheritance

```python
def play(self):
    print("playing")
```

```python
Pet=type('Pet',(),{'play':play})
```

```python
def talk(self):
    print(f"{self.name} says meoww!!")
def c_init(self,name):
    self.name=name
```

```python
Cat=type('Cat',(Animal,Pet),{'__init__':c_init,'sound':talk})
```

```python
multiple=Cat('Kurt')
```

```python
multiple.sound()
multiple.play()
```

### Multilevel inheritance

```python
def fly(self):
    print('flying')
```

```python
Bird=type('Bird',(Animal,),{'fly':fly})
```

```python
def b_init(self,name):
    self.name=name
def speech(self):
    print(f"{self.name} says wee wee")
```

```python
Parrot=type('Parrot',(Bird,),{"__init__":b_init,'sound':speech})
```

```python
multilevel=Parrot('Coco')
```

```python
multilevel.eat()
multilevel.fly()
```

```python
multilevel.sound()
```

### Hybrid inheritance

```python
def walk(self):
    print("walking...")
```

```python
landanimal=type('landanimal',(Animal,),{'walk':walk})
```

```python
def swim(self):
    print("swimmingg....")
```

```python
wateranimal=type('wateranimal',(Animal,),{'swim':swim})
```

```python
def a_init(self,name):
    self.name=name
def a_sound(self):
    print(f"{self.name} speaks..")
```

```python
Amphibians=type('Amphibians',(landanimal,wateranimal),{'__init__':a_init,'sound':a_sound})
```

```python
hybrid=Amphibians('Frog')
```

```python
hybrid.eat()
```

```python
hybrid.walk()
hybrid.swim()
```

```python
hybrid.sound()
```
